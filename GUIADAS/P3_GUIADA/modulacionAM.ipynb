```python
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0, 2, 500)*1e-3 # vector de tiempos (entre 0 y 2 ms)
info = np.cos(2*np.pi*1000*t) # señal de información
port = np.cos(2*np.pi*20000*t) # portadora
m = 0.75 # índice de modulación
modAM = (1 + m*info)*port # señal modulada AM
plt.figure(figsize=(14,8))
plt.subplot(311)
plt.plot(t*1000, info, 'r')
plt.ylabel('Informacion'); plt.grid(axis='y')
plt.xlim(0,2)
plt.subplot(312)
plt.plot(t*1000, port, 'b')
plt.ylabel('Portadora'); plt.grid(axis='y')
plt.xlim(0,2)
plt.subplot(313)
plt.plot(t*1000, modAM, 'm')
plt.plot(t*1000, m*info+1,'--')
plt.ylabel('Señal AM'); plt.grid(axis='y')
plt.xlim(0,2)
plt.xlabel('Tiempo (ms)')
```

```python
plt.figure(figsize=(15,5))
plt.magnitude_spectrum(info, Fs=1/(t[1]-t[0]))
plt.magnitude_spectrum(modAM, Fs=1/(t[1]-t[0]))
plt.legend(['moduladora','modulada'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0, 2, 500)*1e-3 # vector de tiempos (entre 0 y 2 ms)
portadora=np.cos(2*np.pi*20000*t)
moduladora=np.cos(2*np.pi*1000*t)
moduladaAM1=(1+0.75*moduladora)*portadora
t = np.linspace(0, 2, 500)*1e-3 # vector de tiempos (entre 0 y 2 ms)
plt.figure(figsize=(14,8))
plt.subplot(311)
plt.plot(t*1000,moduladaAM1,'r') 
plt.ylabel('AM 20kHz'); plt.grid(axis='y')
plt.xlim(0,2)
portadora=np.cos(2*np.pi*25000*t)
moduladora=np.cos(2*np.pi*1100*t)
moduladaAM2=(1+0.75*moduladora)*portadora
plt.subplot(312)
plt.plot(t*1000, moduladaAM2, 'b') 
plt.ylabel('AM 25kHz'); plt.grid(axis='y')
plt.xlim(0,2)
suma=moduladaAM1+moduladaAM2
plt.subplot(313)
plt.plot(t*1000, suma, 'b') 
plt.ylabel('suma moduladas'); plt.grid(axis='y')
plt.xlim(0,2)
```
